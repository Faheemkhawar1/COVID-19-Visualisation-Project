```python
from Bio import SeqIO
protein_seq=SeqIO.read("protein_seq.fasta","fasta")
```

```python
from Bio.Blast import NCBIWWW
result_handle = NCBIWWW.qblast("blastp", "pdb", protein_seq.seq)
```

```python
from Bio import SearchIO
blast_records = SearchIO.read(result_handle, 'blast-xml')
```

```python
print(blast_records[0:10])
```

```python
# Print all BLAST results
for blast_record in blast_records:
    print(f"Sequence ID: {blast_record.id}")
    print(f"description: {blast_record.description}")
    print(f"E value: {blast_record[0].evalue}")
    print(f"Bit Score:  {blast_record[0].bitscore}")
    print(f"alignment:\n{blast_record.hsps[0].aln}")  # Note the change here
    print("_________________________________________________________________________________")
```

```python
max_bit_score_record = max(blast_records, key=lambda record: record[0].bitscore)

# Print information for the sequence with the highest Bit Score
print("\nInformation for Sequence with Highest Bit Score:")
print(f"Sequence ID: {max_bit_score_record.id}")
print(f"description: {max_bit_score_record.description}")
print(f"E value: {max_bit_score_record[0].evalue}")
print(f"Bit Score:  {max_bit_score_record[0].bitscore}")
print(f"alignment:\n{max_bit_score_record.hsps[0].aln}")  # Note the change here
print("_________________________________________________________________________________")
```
